Numpy



```python
import pandas as pd
import numpy as np

```

```python
df = pd.read_csv("/content/1774404110.csv")

```

```python
df.head()
```

```python
mpg = df['mpg'].to_numpy()


```

```python
mean_mpg = np.mean(mpg)
median_mpg = np.median(mpg)
std_mpg = np.std(mpg)

count_above_25 = np.sum(mpg > 25)

print(mean_mpg, median_mpg, std_mpg, count_above_25)

```

Filtering

```python
cylinders = df['cylinders'].to_numpy()
car_names = df['car name'].to_numpy()

filtered_cars = car_names[cylinders > 6]

filtered_list = filtered_cars.tolist()

print(filtered_list)
```

 Broadcasting

```python
hp = df['horsepower'].to_numpy()

hp = np.where(hp == '?', np.nan, hp).astype(float)

hp_mean = np.nanmean(hp)
hp[np.isnan(hp)] = hp_mean

hp_updated = hp * 1.10

print(hp_updated)
```

4. Boolean Indexing

```python
displacement = df['displacement'].to_numpy()
origin = df['origin'].to_numpy()

avg_disp = np.mean(displacement[origin == 2])

print(avg_disp)
```

Conditional Aggregates (mean mpg by cylinders)

```python
cylinders = df['cylinders'].to_numpy()
mpg = df['mpg'].to_numpy()

unique_cyl = np.unique(cylinders)

mean_mpg_by_cyl = {}

for cyl in unique_cyl:
    mean_mpg_by_cyl[cyl] = np.mean(mpg[cylinders == cyl])

print(mean_mpg_by_cyl)
```

```python
print(df.describe())
```

2

```python
filtered_df = df[(df['model year'] == 75) & (df['weight'] < 3000)]

result = filtered_df[['car name', 'weight', 'mpg']]

print(result)
```

3

```python
print(df.isnull().sum())
```

```python

df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')

median_hp = df['horsepower'].median()
df['horsepower'].fillna(median_hp, inplace=True)

print("Missing values after replacement:")
print(df.isnull().sum())
```

4

```python

df['power_to_weight_ratio'] = df['horsepower'] / df['weight']

print(df[['horsepower', 'weight', 'power_to_weight_ratio']].head())
```

5

```python
grouped = df.groupby('origin')['mpg'].mean()

print(grouped)
```

```python
summary_df = df.groupby("model year")[["mpg", "weight", "horsepower"]].mean()

print(summary_df)
```

```python

high_mpg = df[df["mpg"] > 30][["mpg", "cylinders", "horsepower", "weight"]]

high_mpg.to_csv("high_mpg_cars.csv", index=False)

print("File saved successfully!")
```
